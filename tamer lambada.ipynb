```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("tame.csv")
```

```python
data.drop(["id","uri"],axis=1,inplace=True)

```

```python
data = data.dropna()
```

```python
# change catogrical data to numric
data["album_encoded"]=data["album"].astype('category').cat.codes
data
```

```python
data.drop(["Unnamed: 0","track_number"],axis=1,inplace=True)
corr_matrix = data.corr(numeric_only = True)
print(corr_matrix)
```

```python
plt.figure(figsize=(10, 15))  # Adjust the width and height as needed
top_10=data.sort_values('popularity').tail(10)
colors = ['blue', 'green', 'red', 'purple', 'orange', 'cyan', 'magenta', 'yellow', 'gray', 'teal', 'brown', 'teal', 'lime', 'maroon', 'navy', 'olive', 'aqua', 'fuchsia', 'silver', 'gold']
top_10.plot(x='name',y='popularity',kind='bar',color=colors)
# Add labels and a title
plt.xlabel('Songs')
plt.ylabel('Popularity')
plt.title('Popularity of Songs')
plt.show
```

```python
# List of attributes to plot
attributes_to_plot = ['instrumentalness', 'liveness', 'loudness', 'speechiness', 'tempo', 'valence','popularity']

# Group the data by release_date and calculate the mean for each attribute
grouped_data = data.groupby('release_date')[attributes_to_plot].mean()

# Convert the release_date column to a datetime object (if it's not already)
grouped_data.index = pd.to_datetime(grouped_data.index)

# Create a figure and axis
fig, ax = plt.subplots(figsize=(12, 6))

# Plot lines for each attribute
for attribute in attributes_to_plot:
    grouped_data[attribute].plot(ax=ax, label=attribute, marker='o', linestyle='-', markersize=4)

# Set labels and title
ax.set_xlabel('Release Date')
ax.set_ylabel('Mean Value')
ax.set_title('Mean Attributes Over Time')

# Add a legend
ax.legend()

# Rotate x-axis labels for better readability (optional)
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Import necessary libraries
from dash import Dash, html, dcc, Output, Input
import plotly.express as px
import pandas as pd

# Sample data (you should replace this with your actual data)
data = pd.read_csv('tame.csv')

# Define the Dash app
name = "tame impala dashboard"
app = Dash(__name__)

# Define the layout of the app
app.layout = html.Div(children=[
    html.H1(children='The Tame Impala Dashboard'),

    html.Div(children='''
       An example of a dashboard I'm still learning.
    '''),

    # Add a Dropdown component for song selection
    dcc.Dropdown(
        id='song-selector',
        options=[{'label': song, 'value': song} for song in data['name']],  # Populate dropdown with song names
        value=data['name'].iloc[0],  # Set the initial selected song
        multi=True  # Allow multiple song selection
    ),

    # Add a Dash component to display the bar chart
    dcc.Graph(id='popularity-chart')
])

# Define a callback to update the bar chart based on song selection
@app.callback(
    Output('popularity-chart', 'figure'),
    Input('song-selector', 'value')
)
def update_popularity_chart(selected_songs):
    # Filter the data based on the selected songs
    filtered_data = data[data['name'].isin(selected_songs)]

    # Sort the filtered data by popularity
    sorted_data = filtered_data.sort_values('popularity')

    # Create the bar chart using Plotly Express
    fig = px.bar(sorted_data, x='name', y='popularity', color='name')

    # Customize the layout of the chart
    fig.update_layout(
        xaxis_title='Songs',
        yaxis_title='Popularity',
        title='Popularity of Selected Songs'
    )

    return fig

# Run the Dash app
if __name__ == "__main__":
    app.run_server(debug=True)
```
